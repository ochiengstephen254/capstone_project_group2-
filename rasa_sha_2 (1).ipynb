```python
import rasa
import os

# Train the model
training_result = rasa.train(domain="domain.yml", config="config.yml", training_files=["data/"])

# Extract the latest trained model path
model_directory = "models/"
latest_model = max(
    [os.path.join(model_directory, f) for f in os.listdir(model_directory) if f.endswith('.tar.gz')],
    key=os.path.getctime
)

print(f"Model trained and saved at: {latest_model}")
```

```python
import asyncio
import nest_asyncio
from rasa.core.agent import Agent
from rasa.shared.nlu.training_data.loading import load_data
from sklearn.metrics import accuracy_score, classification_report

# Allow nested asyncio calls (useful in Jupyter, optional if not in Jupyter)
nest_asyncio.apply()

async def evaluate_rasa_nlu_model(model_path, test_data_path):
    try:
        # Load the trained model (synchronous)
        agent = Agent.load(model_path)

        # Load test data with ground truth
        training_data = load_data(test_data_path)
        test_messages = [example.get("text") for example in training_data.nlu_examples]  # Updated to .get("text")
        true_intents = [example.get("intent") for example in training_data.nlu_examples]

        # Predict intents
        predicted_intents = []
        for message in test_messages:
            result = await agent.parse_message(message)
            predicted_intents.append(result["intent"]["name"])

        # Compute metrics
        accuracy = accuracy_score(true_intents, predicted_intents)
        report = classification_report(true_intents, predicted_intents, output_dict=True)

        # Print results
        print(f"Accuracy: {accuracy:.2f}")
        print("Classification Report:")
        print(f"Precision (weighted avg): {report['weighted avg']['precision']:.2f}")
        print(f"Recall (weighted avg): {report['weighted avg']['recall']:.2f}")
        print(f"F1 Score (weighted avg): {report['weighted avg']['f1-score']:.2f}")
        print("\nPer-Intent Metrics:")
        for intent in report.keys():
            if intent not in ["accuracy", "macro avg", "weighted avg"]:
                print(f"Intent: {intent}")
                print(f"  Precision: {report[intent]['precision']:.2f}")
                print(f"  Recall: {report[intent]['recall']:.2f}")
                print(f"  F1 Score: {report[intent]['f1-score']:.2f}")

    except Exception as e:
        print(f"An error occurred: {str(e)}")
```

```python
# TUNED MODEL
# Run the async function
if __name__ == "__main__":
    model_path = "models/20250227-075810-chewy-magpie.tar.gz"
    test_data_path = "data/nlu.yml"
    asyncio.run(evaluate_rasa_nlu_model(model_path, test_data_path))
```
